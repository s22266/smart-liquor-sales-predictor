```python
import seaborn as sns

outliers_dict = {}

for column in df.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    outliers_dict[column] = outliers
    
sns.set(style="whitegrid")
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Outliners for column: {column}')
    plt.show()
```

```python
selected_columns = ['store', 'zipcode', 'vendor_no', 'itemno',
                    'state_bottle_cost', 'state_bottle_retail', 'sale_bottles',
                    'sale_dollars', 'sale_liters', 'sale_gallons']

outliers_dict = {}

num_rows = (len(selected_columns) + 2) // 3 
num_cols = min(len(selected_columns), 3)  

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))

for i, column in enumerate(selected_columns):
    row = i // num_cols
    col = i % num_cols

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    outliers_dict[column] = outliers

    # Deleting outliers for specific column
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    sns.boxplot(x=df[column], ax=axes[row, col])
    axes[row, col].set_title(f'Outliners for column: {column}')

plt.tight_layout()
plt.show()
```

FEATURE ENGINEERING

1. NEW FEATURES CREATING

```python
from datetime import datetime

def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'
    
def get_weekday_info(year, month, day):
    date = datetime(year, month, day)
    day_of_week = date.weekday()
    is_weekday = day_of_week < 5
    return day_of_week, is_weekday
```

```python
df['day_of_week'], df['is_weekday'] = zip(*df.apply(lambda row: get_weekday_info(row['year'], row['month'], row['day']), axis=1))
df['season'] = df['month'].apply(get_season)
```
